```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
image_path = 'DIP_Lab_task_04.jpg'
image = cv2.imread(image_path)
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
brightness_reduced_image = cv2.convertScaleAbs(image, alpha=1, beta=-50)
```

```python
brightness_reduced_image_rgb = cv2.cvtColor(brightness_reduced_image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(brightness_reduced_image_rgb)
plt.title('Brightness Reduced Image')
plt.axis('off')
plt.show()
```

```python
cv2.imwrite('brightness_reduced_image.jpg', brightness_reduced_image)
```
