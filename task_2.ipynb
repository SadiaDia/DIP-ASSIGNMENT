```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
image = Image.open('DIP_Lab_task_02.jpg')
```

```python
gray_image = image.convert('L')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].imshow(image)
axes[0].set_title('Original RGB Image')
axes[0].axis('off')
axes[1].imshow(gray_image, cmap='gray')
axes[1].set_title('Grayscale Image')
axes[1].axis('off')
plt.show()
```

```python
rgb_array = np.array(image)
gray_array = np.array(gray_image)
```

```python
y = 100
x = 100
```

```python
rgb_subregion = rgb_array[y:y+5, x:x+5]
gray_subregion = gray_array[y:y+5, x:x+5]
```

```python
print("RGB Sub-region (5x5):")
print(rgb_subregion)
print("\nGrayscale Sub-region (5x5):")
print(gray_subregion)
```
