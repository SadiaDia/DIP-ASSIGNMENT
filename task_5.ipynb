```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
image_path = 'DIP_Lab_task_05.jpg'
image = cv2.imread(image_path)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```python
ret, otsu_segmented_image = cv2.threshold(gray_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
plt.imshow(otsu_segmented_image, cmap='gray')
plt.title('Otsu Segmented Image')
plt.axis('off')
plt.show()
```

```python
cv2.imwrite('otsu_segmented_image.jpg', otsu_segmented_image)
```
